```python
X['Sex'] = X['Sex'].map({'male': 1, 'female': 0})
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
age_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer
categorical_features = ['Embarked', 'Cabin']
```

```python
from sklearn.compose import ColumnTransformer
categorical_features = ['Embarked', 'Cabin']
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),  
    ('encoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('age', age_pipeline, ['Age']),               
    ('cat', categorical_pipeline, categorical_features)  
], remainder='passthrough')
```

```python
X_transformed = preprocessor.fit_transform(X)
```

```python
X_transformed
```

```python
age_cols = ['Age']

cat_cols = preprocessor.named_transformers_['cat']['encoder'].get_feature_names_out(categorical_features)

remainder_cols = [col for col in X.columns if col not in ['Age'] + categorical_features]
all_columns = list(age_cols) + list(cat_cols) + remainder_cols

X_df = pd.DataFrame(X_transformed, columns=all_columns)
X_df
```

```python
X_df.columns = X_df.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_df,y,test_size = 0.2,random_state = 40)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42,n_estimators = 500,criterion = 'entropy',min_samples_split= 5,max_features = 'sqrt',oob_score= True)
rf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(rf,X_train,y_train,cv = 5,scoring = 'accuracy').mean()
```

```python
y_predicted = rf.predict(X_test)
y_predicted
```

```python
importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending = False)
importances
```

```python
importances.to_csv('feature_importance.csv')
```

```python
top_n = 25
top_features = importances.head(top_n).index.tolist()
print(f"Top {top_n} features used for retraining:\n", top_features)
```

```python
corr_with_target = X_df[top_features].corrwith(y)
corr_with_target.sort_values(ascending=False).plot(kind='bar', figsize=(10,4))
plt.title('Correlation of Features with Survival')
plt.show()
```

```python
top_features_importance = importances.head(20)

plt.figure(figsize=(12, 8))

sns.barplot(
    x=top_features_importance.values,
    y=top_features_importance.index,
    hue=top_features_importance.index,  
    palette='crest',
    dodge=False,                        
    legend=False                        
)

plt.title('Top 20 Most Important Features (Random Forest)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

display_labels = ['Not Survived', 'Survived']

cm_disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
cm_disp.plot(cmap = 'Blues')
plt.title('Confusion Matrix for Random Forest')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_test,y_scores[:,1])
plt.figure(figsize=(8, 6))
plt.plot(recalls, precisions, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```

```python
target_value = 0.82
idx = np.argmin(np.abs(precisions - target_value) + np.abs(recalls - target_value))
best_threshold = thresholds[idx]
print(f"Best threshold for precisionâ‰ˆrecallâ‰ˆ80%: {best_threshold:.4f}")
print(f"Precision: {precisions[idx]:.4f}, Recall: {recalls[idx]:.4f}")
```

```python
y_pred_custom = (y_scores >= best_threshold).astype(int)
y_pred_custom
```

```python
pd.DataFrame(y_pred_custom[:,1]).value_counts()
```

```python
precision = precision_score(y_test, y_pred_custom[:,1])
recall = recall_score(y_test, y_pred_custom[:,1])
f1 = f1_score(y_test, y_pred_custom[:,1])
roc_auc = roc_auc_score(y_test, y_pred_custom[:,1])
print(f"Test Accuracy: {accuracy_score(y_test, y_pred_custom[:,1])*100:.2f}%")
print(f"Precision score: {(precision)*100:.2f}%")
print(f"Recall score: {(recall)*100:.2f}%")
print(f"f1-score: {(f1)*100:.2f}%")
print(f"ROC-AUC score: {(roc_auc)*100:.2f}%")
```

```python
cm_final = confusion_matrix(y_test,y_pred_custom[:,1])
cm_final
```

```python
cm_disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=display_labels)
cm_disp_final.plot(cmap = 'Blues')
plt.title('Confusion Matrix After Optimisation')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr,threshold = roc_curve(y_test,y_scores[:,1])
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.3f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic(ROC)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred_custom[:,1], output_dict=True)
clf_report = pd.DataFrame(clf_report).transpose()
clf_report = clf_report.drop(columns=['support'])
plt.figure(figsize=(8, 6))
sns.heatmap(clf_report.iloc[:-3, :].T, annot=True, cmap='BuGn', fmt=".2f")
plt.title('Classification Report Heatmap')
plt.ylabel('Metrics--->')
plt.xlabel('Classes--->')
plt.show()
```

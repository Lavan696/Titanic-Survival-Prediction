```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df['FamilySize'] = df['Parch'] + df['SibSp'] + 1
df
```

```python
df['FarePerPerson'] = df['Fare']/df['FamilySize']
df
```

```python
df.drop(['Name','SibSp','Parch','Ticket'],inplace = True,axis = 1)
df
```

```python
y = df['Survived']
y
```

```python
X = df.drop('Survived',axis = 1)
X
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'median')
X[['Age']] = imputer.fit_transform(X[['Age']])
X
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()


age_scale = X[['Age']] 

X['Age'] = std_scaler.fit_transform(age_scale)
X
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output= False)
```

```python
encoder = OneHotEncoder(sparse_output=False)
embarked_encoded = encoder.fit_transform(X[['Embarked']])
embarked_encoded = pd.DataFrame(
    embarked_encoded,
    columns=encoder.get_feature_names_out(['Embarked'])
)
```

```python
X = pd.concat([X,embarked_encoded],axis = 1)
X
```

```python
X.drop('Embarked',axis = 1,inplace = True)
X
```

```python
X['Sex'] = X['Sex'].map({'male': 1, 'female': 0})
```

```python
X
```

```python
cabin = X['Cabin']
X.drop('Cabin',axis = 1,inplace = True)
cabin_encoded = pd.DataFrame(encoder.fit_transform(pd.DataFrame(cabin)),columns = encoder.categories_) 
cabin_encoded
```

```python
X = pd.concat([X,cabin_encoded],axis = 1)
X
```

```python
X.columns = X.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 40)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42,n_estimators = 500,criterion = 'entropy',min_samples_split= 5,max_features = 'sqrt',oob_score= True)
rf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(rf,X_train,y_train,cv = 5,scoring = 'accuracy').mean()
```

```python
y_predicted = rf.predict(X_test)
y_predicted
```

```python
importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending = False)
importances
```

```python
top_n = 20
top_features = importances.head(top_n).index.tolist()
print(f"Top {top_n} features used for retraining:\n", top_features)
```

```python
X_train_top , X_test_top = X_train[top_features],X_test[top_features]
```

```python
rf.fit(X_train_top,y_train)
```

```python
y_predicted  = rf.predict(X_test_top)
y_predicted
```

```python
y_scores = rf.predict_proba(X_test_top)
y_scores
```

```python
rf_cvs = cross_val_score(rf,X_train_top,y_train,cv = 5,scoring = 'accuracy')
rf_cvs.mean()
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr,threshold = roc_curve(y_test,y_scores[:,1])
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.3f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic(ROC)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
target_value = 0.80
idx = np.argmin(np.abs(precisions - target_value) + np.abs(recalls - target_value))
best_threshold = thresholds[idx]
print(f"Best threshold for precision≈recall≈80%: {best_threshold:.4f}")
print(f"Precision: {precisions[idx]:.4f}, Recall: {recalls[idx]:.4f}")
```

```python
y_pred_custom = (y_scores >= best_threshold).astype(int)
y_pred_custom
```

```python
precision = precision_score(y_test, y_pred_custom[:,1])
recall = recall_score(y_test, y_pred_custom[:,1])
f1 = f1_score(y_test, y_pred_custom[:,1])
roc_auc = roc_auc_score(y_test, y_pred_custom[:,1])
print(f"Accuracy: {accuracy_score(y_test, y_pred_custom[:,1]):.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
```
